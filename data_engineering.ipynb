```python
# Import dependencies
import pandas as pd
import matplotlib as plt
import numpy as np
import os
```

```python
# Read in the measurement CSV file as DataFrame; inspect for missing data
hawaii_measurements = pd.read_csv("hawaii_measurements.csv")
hawaii_measurements.head(50)
```

```python
# Drop rows with missing data
hawaii_measurements = hawaii_measurements.dropna(axis=0)
hawaii_measurements.head()
```

```python
# Reset index after removing null values
clean_hawaii_measurements = hawaii_measurements.reset_index(drop=True)
clean_hawaii_measurements.head()
```

```python
# Save clean csv file
cleaned_hawaii_measurements = "clean_hawaii_measurements_csv.csv"
clean_hawaii_measurements.to_csv(cleaned_hawaii_measurements, index=False)
```
